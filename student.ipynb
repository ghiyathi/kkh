```python
# Python program for implementation of Quicksort Sort 

# This function takes first element as pivot, places 
# the pivot element at its correct position in sorted 
# array, and places all smaller (smaller than pivot) 
# to left of pivot and all greater elements to right 
# of pivot 

def quicksort(arr):
    if len(arr)<2:
        return arr
    
    left,same,right = [],[],[]
    pivot = arr[0]
    
    for item in arr:
        if item < pivot:
            left.append(item)
        elif item == pivot:
            same.append(item)
        elif item > pivot:
            right.append(item)
    return quicksort(left) + same + quicksort(right)

# code to test above 
arr = [10, 7, 8, 9, 1, 5] 
print("Sorted array is:{}".format(quicksort(arr)))


```
